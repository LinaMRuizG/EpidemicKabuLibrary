```python
#step1: building the groups
df.loc[:,"cutDatesW2"]=df["cutDatesW"].cumsum()
df.head(3)
```

```python
#step2: getting the peaks dates
peaksDates = df.groupby(df["cutDatesW2"]).apply(
    lambda x :
    x[x[casesName]==max(x[casesName])][datesName]
)
print(type(peaksDates))
print(peaksDates[0])
peaksDates
```

```python
#step2: crating the peaksDates column in the df
df["peaksDates"] = df[datesName].isin(peaksDates).astype(int)
df.head(2)
```

```python
from datetime import Timedelta

data = [
    (Timestamp('2020-07-22 00:00:00'), Timedelta('-98 days +00:00:00')),
    (Timestamp('2020-08-14 00:00:00'), Timedelta('-75 days +00:00:00')),
    (Timestamp('2020-10-08 00:00:00'), Timedelta('-20 days +00:00:00')),
    (Timestamp('2020-11-11 00:00:00'), Timedelta('14 days 00:00:00'))
]

# Convert Timedeltas to days for comparison
timedeltas_days = [td.days for _, td in data]

# Find the indices of the lowest negative and lowest positive Timedeltas
lowest_negative_index = timedeltas_days.index(min(td for td in timedeltas_days if td < 0))
lowest_positive_index = timedeltas_days.index(min(td for td in timedeltas_days if td > 0))

# Get the tuples corresponding to the lowest negative and lowest positive Timedeltas
tuple_lowest_negative = data[lowest_negative_index]
tuple_lowest_positive = data[lowest_positive_index]

print("Tuple with Lowest Negative Timedelta:", tuple_lowest_negative)
print("Tuple with Lowest Positive Timedelta:", tuple_lowest_positive)
```
